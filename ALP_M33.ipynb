3 - Crie um algoritmo, que receba o nome do usuário, e 3 notas, depois calcule a média na variável (média), e imprima na tela a seguinte frase:

O aluno (nome) está com (média) de média!

```python
nome = str(input("Digite seu nome: "))
Nota1 = float(input("Digite sua nota: "))
Nota2 = float(input("Digite sua nota: "))
Nota3 = float(input("Digite sua nota: "))

#entrada

média = (Nota1 + Nota2 + Nota3) / 3

#saída
print("O aluno", nome, "está com", média, "de média!")
print(f"O aluno {nome} está com {média:.2f} de média!")
```

## Biblioteca Random

Gera valores aleatórios

```python
import random
```

```python
alt = random.randint(0, 100)
tentativa = 5

while (tentativa>0):
  numero = int(input("Digite um número entre 0 e 100: "))
  if (numero==alt):
    print("Parabéns! Você acertou!!!")
    break
  elif (numero<alt):
    print("Digite um número maior!")
  else:
    print("Digite um número menor!")
  tentativa -=1
if (tentativa == 0):
  print("Suas tentativas acabaram! Você não acertou!")
  print("O número era: ", alt)
```

```python
while True:
    print("=========================")
    print("   CALCULADORA INTERATIVA")
    print("=========================")
    print("Digite o número da opção desejada:")
    print("1 - Múltiplos de 5 em um Intervalo")
    print("2 - Contagem de Números Pares em um Intervalo")
    print("3 - Verificador de Número Primo")
    print("4 - Conversor de Temperatura (Cº para Fº)")
    print("0 - Sair do Programa")
    print("=========================")

    opcao = input("Escolha uma opção: ")

    # --- OPÇÃO 0: SAIR ---
    if opcao == "0":
        print("Saindo do programa... Até a próxima!")
        break

    # --- OPÇÃO 1: MÚLTIPLOS DE 5 ---
    elif opcao == "1":
        inicio = int(input("Digite o número inicial: "))
        fim = int(input("Digite o número final: "))

        print(f"\nMúltiplos de 5 entre {inicio} e {fim}:")
        for i in range(inicio, fim + 1):
            if i % 5 == 0:
                print(i)
        print()

    # --- OPÇÃO 2: CONTAGEM DE NÚMEROS PARES ---
    elif opcao == "2":
        inicio = int(input("Digite o número inicial: "))
        fim = int(input("Digite o número final: "))

        cont = 0
        for i in range(inicio, fim + 1):
            if i % 2 == 0:
                cont += 1

        print(f"\nQuantidade de números pares entre {inicio} e {fim}: {cont}\n")

    # --- OPÇÃO 3: VERIFICADOR DE NÚMERO PRIMO ---
    elif opcao == "3":
        num = int(input("Digite um número inteiro positivo: "))

        if num <= 1:
            print(f"\nO número {num} NÃO é primo.\n")
        else:
            primo = True
            for i in range(2, num):
                if num % i == 0:
                    primo = False
                    break

            if primo:
                print(f"\nO número {num} é PRIMO.\n")
            else:
                print(f"\nO número {num} NÃO é primo.\n")

    # --- OPÇÃO 4: CONVERSOR DE TEMPERATURA ---
    elif opcao == "4":
        c = float(input("Digite a temperatura em Celsius: "))
        f = c * 1.8 + 32
        print(f"\n{c}°C corresponde a {f:.2f}°F\n")

    # --- OPÇÃO INVÁLIDA ---
    else:
        print("\nOpção inválida! Tente novamente.\n")
```
